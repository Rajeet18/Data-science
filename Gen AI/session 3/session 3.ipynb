1.Write a Python function called simple_generator() that takes a random vector as input and returns a 28x28 array of grayscale pixel values between 0 and 1, simulating a fake handwritten digit image like those used in GAN demos.

```python
import numpy as np

def simple_generator(random_vector):
    """
    Generates a fake 28x28 grayscale image
    from a random input vector.
    """
    
    # Ensure vector has 784 values (28x28)
    if len(random_vector) != 784:
        random_vector = np.resize(random_vector, 784)

    # Apply sigmoid activation to scale values between 0 and 1
    image = 1 / (1 + np.exp(-random_vector))

    # Reshape into 28x28 image
    image = image.reshape(28, 28)

    return image

# Example usage
latent_vector = np.random.randn(100)
fake_digit = simple_generator(latent_vector)

print("Shape:", fake_digit.shape)
print("Pixel range:", fake_digit.min(), "to", fake_digit.max())
```

2.Create a Python class Discriminator with a method discriminate(image) that takes a 28x28 image array and returns a probability score between 0 and 1 indicating if the image is 'real' or 'fake'.<br><br><em><strong>

```python
import numpy as np

class Discriminator:
    def __init__(self, threshold=0.5):
        self.threshold = threshold

    def discriminate(self, image):
        """
        Returns a probability score between 0 and 1
        indicating whether the image is real.
        """
        mean_pixel = np.mean(image)

        # Simple probability based on threshold
        probability = min(mean_pixel / self.threshold, 1.0)

        return probability

image = np.random.rand(28, 28)  # Random grayscale image

discriminator = Discriminator(threshold=0.5)

score = discriminator.discriminate(image)

print("Realness Score:", score)

if score > 0.5:
    print("Prediction: Real")
else:
    print("Prediction: Fake")
```

3.Simulate one cycle of the adversarial training loop: generate a fake image using your simple_generator(), get a score from your Discriminator, and print whether the Discriminator correctly identifies the fake image.

```python
import numpy as np

def simple_generator(random_vector):
    if len(random_vector) != 784:
        random_vector = np.resize(random_vector, 784)

    image = 1 / (1 + np.exp(-random_vector))  # Sigmoid
    return image.reshape(28, 28)

class Discriminator:
    def __init__(self, threshold=0.5):
        self.threshold = threshold

    def discriminate(self, image):
        mean_pixel = np.mean(image)
        probability = min(mean_pixel / self.threshold, 1.0)
        return probability

latent_vector = np.random.randn(100)
fake_image = simple_generator(latent_vector)

discriminator = Discriminator(threshold=0.5)
score = discriminator.discriminate(fake_image)

print("Discriminator Score:", round(score, 3))

if score > 0.5:
    prediction = "Real"
else:
    prediction = "Fake"

print("Prediction:", prediction)

if prediction == "Fake":
    print("✅ Discriminator correctly identified the fake image.")
else:
    print("❌ Discriminator failed and thought the fake image was real.")
```
