```python
#8. Check whether a list contains a sublist
def contains_sublist(lst, sub):
    for i in range(len(lst) - len(sub) + 1):
        if lst[i:i+len(sub)] == sub:
            return True
    return False

print(contains_sublist([1,2,3,4,5], [4,5]))
```

```python
#17. Sum of odd and even series

n = int(input("Enter a number (N): "))

sum_even = 0
sum_odd = 0

for num in range(1, n + 1):
    if num % 2 == 0:
        sum_even += num
    else:
        sum_odd += num

print(f"The sum of even numbers from 1 to {n} is: {sum_even}")
print(f"The sum of odd numbers from 1 to {n} is: {sum_odd}")
```

```python
#18. Python Program to Find Factorial of Number Using Recursion
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

num = 5
print(f"Factorial of {num} is {factorial(num)}")
```

```python
#19. Python Function to Return Unique Elements from a List
def unique_elements(lst):
    unique_lst = []
    for item in lst:
        if item not in unique_lst:
            unique_lst.append(item)
    return unique_lst

my_list = [1, 2, 2, 3, 4, 4, 5]
print("Original list:", my_list)
print("Unique elements:", unique_elements(my_list))
```

```python
#20.Mini Project: Password Generator
import random
import string

def generate_password():
    user_input = input("Enter some words separated by spaces: ").split()
    
    if not user_input:
        return "No words provided!"
    
    password_words = ''.join(random.sample(user_input, min(2, len(user_input))))
    
    password_chars = (
        random.choice(string.ascii_uppercase) +
        random.choice(string.digits) +
        random.choice(string.punctuation)
    )

    password = password_words + password_chars
    
    while len(password) < 8:
        password += random.choice(string.ascii_letters + string.digits + string.punctuation)
    
    password = ''.join(random.sample(password, len(password)))
    
    return password

print("Generated Password:", generate_password())
```
