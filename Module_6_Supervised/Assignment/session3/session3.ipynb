## Q2


```python
import pandas as pd

ipl = pd.read_csv("ipl_sample.csv")

median_age = ipl["player_age"].median()

ipl["player_age"] = ipl["player_age"].fillna(median_age)

print("Updated player_age column:")
print(ipl["player_age"])
```

## Q5


```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

ipl = pd.read_csv("ipl_sample.csv")

encoder = LabelEncoder()

ipl["player_role_encoded"] = encoder.fit_transform(ipl["player_role"])

print(ipl[["player_role", "player_role_encoded"]])

print("\nRole Mapping:")

mapping = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))

print(mapping)
```
