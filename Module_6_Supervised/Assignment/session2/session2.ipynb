## Q-1

```python
import pandas as pd

df = pd.read_csv("data.csv")  

print(df.head())

print(df.columns)
```

```python
features = [
    'danceability',
    'energy',
    'key',
    'loudness',
    'speechiness',
    'acousticness',
    'instrumentalness',
    'liveness',
    'valence',
    'tempo',
    'duration_ms'
]
```

```python
label = "popularity"
```

## Q2


```python
from sklearn.model_selection import train_test_split

X = df[features]

y = df[label]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training rows:", len(X_train))
print("Testing rows :", len(X_test))
```

## Q3

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X = df[['danceability']]
y = df['popularity']

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.05,test_size=0.95,random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

train_pred = model.predict(X_train)
test_pred = model.predict(X_test)

print("Train R²:", r2_score(y_train, train_pred))
print("Test R² :", r2_score(y_test, test_pred))

print("Train MSE:", mean_squared_error(y_train, train_pred))
print("Test MSE :", mean_squared_error(y_test, test_pred))
```

Short Note

In a Zomato restaurant rating predictor, a model with high bias may predict nearly the same rating for every restaurant, ignoring important factors like food quality and customer reviews. A model with high variance may memorize past ratings and fail to predict ratings accurately for new restaurants. A good model balances bias and variance to provide accurate predictions for both existing and new restaurants.

Example: Instagram Reels

Suppose Instagram recommends Reels based only on the last few videos you watched. If you watch several cricket videos in one day, the recommendation system may start showing almost only cricket content. This is an example of overfitting, because the model has learned too much from a small set of recent interactions and ignores your broader interests such as music, travel, or technology.
