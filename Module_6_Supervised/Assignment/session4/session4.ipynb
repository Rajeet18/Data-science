## Q1

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

ipl = pd.DataFrame({
    "runs": [450, 380, 510, 275, 620],
    "wickets": [5, 12, 3, 18, 1],
    "matches": [14, 13, 15, 12, 16],
    "strike_rate": [145.6, 132.8, 151.2, 128.5, 160.4]
})

ipl.to_csv("ipl_players.csv", index=False)

df = pd.read_csv("ipl_players.csv")

scaler = StandardScaler()

numeric_cols = ["runs", "wickets", "matches", "strike_rate"]

df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

df.to_csv("ipl_scaled.csv", index=False)

print(df)
print("\nScaled data saved as 'ipl_scaled.csv'")
```

## Q2

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

flipkart = pd.DataFrame({
    "price": [1200, 2500, 1800, 3200, 1500],
    "rating": [4.2, 3.8, 4.5, 4.0, 4.7],
    "reviews": [120, 340, 560, 230, 450],
    "discount": [10, 25, 15, 30, 20]
})

print("Before Scaling")

print(flipkart.min())
print()
print(flipkart.max())

scaler = MinMaxScaler()

scaled = pd.DataFrame(
    scaler.fit_transform(flipkart),
    columns=flipkart.columns
)

print("\nAfter Scaling")

print(scaled.min())
print()
print(scaled.max())
```

## Q3

```python
import pandas as pd

spotify = pd.DataFrame({
    "danceability": [0.82,0.75,0.60,0.91,0.66,0.88],
    "energy": [0.90,0.65,0.70,0.95,0.55,0.92],
    "tempo": [120,110,130,125,105,128],
    "popularity": [95,80,70,98,60,96]
})

corr = spotify.corr()

print("Correlation Matrix\n")
print(corr)

target_corr = corr["popularity"].drop("popularity")

top2 = target_corr.abs().sort_values(ascending=False).head(2)

print("\nTop 2 Features:")

for feature in top2.index:
    print(feature)
```

## Q4


```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.feature_selection import RFE

zomato = pd.DataFrame({
    "rating":[4.5,4.2,3.8,4.7,4.1],
    "votes":[500,250,180,650,300],
    "delivery_time":[30,45,35,25,40],
    "distance":[2.5,5.2,3.1,1.8,4.0],
    "average_cost_for_two":[1200,800,600,1500,900]
})

X = zomato.drop("average_cost_for_two", axis=1)
y = zomato["average_cost_for_two"]

model = DecisionTreeRegressor(random_state=42)

selector = RFE(model, n_features_to_select=3)

selector.fit(X, y)

selected = X.columns[selector.support_]

print("Selected Features:")

for feature in selected:
    print(feature)

print("\nRFE Ranking")

for col, rank in zip(X.columns, selector.ranking_):
    print(col, ":", rank)
```

## Q5

```python
import pandas as pd

movies = pd.DataFrame({
    "movie": ["A","B","C","D","E","F","G","H"],
    "user_ratings": [4.2,4.5,4.6,4.4,4.8,4.3,9.8,1.2]
})

Q1 = movies["user_ratings"].quantile(0.25)

Q3 = movies["user_ratings"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR

upper = Q3 + 1.5 * IQR

clean_data = movies[
    (movies["user_ratings"] >= lower) &
    (movies["user_ratings"] <= upper)
]

rows_removed = len(movies) - len(clean_data)

print("Original Rows:", len(movies))
print("Rows After Removing Outliers:", len(clean_data))
print("Rows Dropped:", rows_removed)

print("\nClean Dataset")

print(clean_data)
```
