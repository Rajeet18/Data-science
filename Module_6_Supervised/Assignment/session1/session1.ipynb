Traditional programming would detect spam by using predefined rules. For example, if a message contains words like "Win Money" or includes too many links, it is marked as spam. However, these rules must be manually created and updated whenever new spam patterns appear.

Machine learning, on the other hand, learns from a large dataset of labeled messages (spam and non-spam). Instead of relying on fixed rules, it identifies patterns automatically and predicts whether a new message is spam. As more data becomes available, the model can improve its accuracy and adapt to new types of spam without requiring manual rule updates.
