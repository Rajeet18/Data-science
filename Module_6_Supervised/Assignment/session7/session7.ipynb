## Q1

```python
import numpy as np
from sklearn.metrics import mean_squared_error

actual = np.array([30, 25, 40, 35, 28, 50, 45, 38, 32, 27])
predicted = np.array([32, 24, 39, 36, 30, 48, 47, 37, 31, 29])

mse = mean_squared_error(actual, predicted)

rmse = np.sqrt(mse)

print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

## Q2

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, r2_score

actual_prices = np.array([15000, 18000, 22000, 25000, 30000])
predicted_prices = np.array([14800, 18200, 21800, 25200, 29500])

def evaluate(actual, predicted):
    mae = mean_absolute_error(actual, predicted)
    r2 = r2_score(actual, predicted)

    print("Mean Absolute Error (MAE):", mae)
    print("R² Score:", r2)

evaluate(actual_prices, predicted_prices)
```

## Q3

```python
import numpy as np
import matplotlib.pyplot as plt

actual = np.array([
    4.2, 3.8, 4.5, 2.9, 4.8,
    3.5, 4.1, 2.7, 3.9, 4.6,
    4.0, 3.2, 4.7, 2.8, 3.6,
    4.3, 3.7, 4.9, 2.5, 3.4
])

predicted = np.array([
    4.1, 3.9, 4.4, 3.0, 4.7,
    3.6, 4.0, 2.9, 4.0, 4.5,
    3.9, 3.3, 4.6, 2.7, 3.5,
    4.4, 3.8, 4.8, 2.6, 3.3
])

residuals = actual - predicted

plt.figure(figsize=(8,5))

plt.scatter(predicted, residuals, color='blue')

plt.axhline(y=0, color='red', linestyle='--')

plt.xlabel("Predicted Ratings")
plt.ylabel("Residuals")

plt.title("Residual Plot")

plt.grid(True)

plt.show()
```

## Q5

```python
import numpy as np
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    r2_score
)

actual = np.array([70, 80, 65, 90, 75, 85, 95, 60])

pred_a = np.array([72, 78, 66, 88, 76, 84, 93, 61])

pred_b = np.array([68, 82, 60, 94, 72, 87, 98, 55])

def evaluate_model(name, actual, predicted):
    mse = mean_squared_error(actual, predicted)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(actual, predicted)
    r2 = r2_score(actual, predicted)

    print(f"\n{name}")
    print("MSE :", mse)
    print("RMSE:", rmse)
    print("MAE :", mae)
    print("R²  :", r2)

evaluate_model("Model A", actual, pred_a)
evaluate_model("Model B", actual, pred_b)
```
