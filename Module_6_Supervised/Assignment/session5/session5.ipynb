## Q1

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

instagram_hours = np.array([1, 2, 3, 4, 5, 6, 7])

battery_percentage = np.array([90, 84, 78, 70, 65, 55, 48])

plt.figure(figsize=(8,5))
plt.scatter(instagram_hours, battery_percentage,
            color='blue', s=80)

plt.title("Instagram Usage vs Battery Percentage")
plt.xlabel("Instagram Usage (Hours)")
plt.ylabel("Battery Remaining (%)")
plt.grid(True)

plt.show()
```

## Q2


```python
X = instagram_hours.reshape(-1, 1)
y = battery_percentage

model = LinearRegression()

model.fit(X, y)

print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

## Q3

```python
predicted = model.predict(X)

plt.figure(figsize=(8,5))
plt.scatter(instagram_hours, battery_percentage,
            color='blue', label="Actual Data")

plt.plot(instagram_hours, predicted,
         color='red', linewidth=2,
         label="Regression Line")

plt.title("Linear Regression")
plt.xlabel("Instagram Usage (Hours)")
plt.ylabel("Battery Remaining (%)")
plt.legend()
plt.grid(True)

plt.show()
```

## Q4

```python
mse = mean_squared_error(y, predicted)

print("\nMean Squared Error (MSE):", mse)

print("\nExplanation:")
print("A lower MSE indicates that the predicted values are closer to the actual values, meaning the regression model fits the data better.")
```
